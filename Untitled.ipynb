```python
 # Use a custom color palette
custom_palette = sns.color_palette("Set2")  # Adjust the palette as needed
plt.figure(figsize=(14, 8))
sns.relplot(x="Country/Region", y="Active", hue="WHO Region", data=data, kind="scatter", palette=custom_palette)
plt.xlabel("Country/Region")
plt.ylabel("Active Cases")
plt.title("Scatter Plot of Active Cases by Country/Region and WHO Region")
plt.xticks(rotation=45, ha="right")
plt.show()
```

```python
# Use a different plot type, e.g., a bar plot
plt.figure(figsize=(14, 8))
sns.barplot(x="Country/Region", y="Active", hue="WHO Region", data=data)
plt.xlabel("Country/Region")
plt.ylabel("Active Cases")
plt.title("Bar Plot of Active Cases by Country/Region and WHO Region")
plt.xticks(rotation=45, ha="right")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.relplot(x="Country/Region", y="Active", hue="WHO Region", data=data, kind="scatter")
plt.xlabel("Country/Region")
plt.ylabel("Active Cases")
plt.title("Scatter Plot of Active Cases by Country/Region and WHO Region")
plt.xticks(rotation=45, ha="right")  # Rotate x-axis labels for better readability
plt.show()
```

```python
# Assuming 'Date' is in datetime format
data['Date'] = pd.to_datetime(data['Date'])

# Time series plot of confirmed cases over time
plt.figure(figsize=(12, 6))
sns.lineplot(x='Date', y='Confirmed', data=data, hue='Country/Region')
plt.title('Time Series of Confirmed Cases')
plt.xlabel('Date')
plt.ylabel('Confirmed Cases')
plt.show()
```

```python
import geopandas as gpd
from mpl_toolkits.axes_grid1 import make_axes_locatable

# Assuming 'Lat' and 'Long' columns are present
gdf = gpd.GeoDataFrame(data, geometry=gpd.points_from_xy(data['Long'], data['Lat']))

world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

fig, ax = plt.subplots(figsize=(15, 10))
divider = make_axes_locatable(ax)
cax = divider.append_axes("right", size="5%", pad=0.1)

world.plot(ax=ax, color='lightgrey')
gdf.plot(ax=ax, color='red', markersize=10, label='COVID-19 Cases', alpha=0.7)
ax.legend()
plt.title('Geographical Distribution of COVID-19 Cases')
plt.show()
```

```python
top_countries_confirmed = data.groupby('Country/Region')['Confirmed'].max().sort_values(ascending=False).head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=top_countries_confirmed.index, y=top_countries_confirmed.values)
plt.title('Top 10 Countries with the Highest Confirmed Cases')
plt.xlabel('Country/Region')
plt.ylabel('Confirmed Cases')
plt.xticks(rotation=45, ha="right")
plt.show()
```

```python
data['Mortality Rate'] = data['Deaths'] / data['Confirmed']
data['Recovery Rate'] = data['Recovered'] / data['Confirmed']

plt.figure(figsize=(12, 6))
sns.lineplot(x='Date', y='Mortality Rate', data=data, hue='Country/Region')
plt.title('Mortality Rate Over Time')
plt.xlabel('Date')
plt.ylabel('Mortality Rate')
plt.show()
```
